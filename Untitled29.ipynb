```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
# Prepare data

# Encode target
df1["Churn"] = df1["Churn"].replace({"Yes": 1, "No": 0})
```

```python
# Encode target
df1["Churn"] = df1["Churn"].replace({"Yes": 1, "No": 0})
```

```python
# One-hot encode categorical columns
df1 = pd.get_dummies(df1, drop_first=True)
```

```python
# Separate features & target
X = df1.drop("Churn", axis=1)
y = df1["Churn"]
```

```python
# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df1
```

```python
# Clustering

# Find optimal cluster - Elbow method

inertia = []
for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 8), inertia, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
# Apply K means

kmeans = KMeans(n_clusters=3, random_state=42)
df1["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# Second clustering algorithm - hierarchical

hierarchical = AgglomerativeClustering(n_clusters=3)
df1["Hierarchical_Cluster"] = hierarchical.fit_predict(X_scaled)
```

```python
df1["Hierarchical_Cluster"]
```

```python
# Segment Analysis

segment_summary = df1.groupby("Cluster").mean()
segment_summary
```

```python
# Build prediction model per segment

results = []

for segment in df1["Cluster"].unique():
    segment_data = df1[df1["Cluster"] == segment]

    X_seg = segment_data.drop(["Churn", "Cluster", "Hierarchical_Cluster"], axis=1)
    y_seg = segment_data["Churn"]

    X_train, X_test, y_train, y_test = train_test_split(
        X_seg, y_seg, test_size=0.2, random_state=42
    )

    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)

    preds = rf.predict(X_test)
    probs = rf.predict_proba(X_test)[:,1]

    results.append([
        segment,
        accuracy_score(y_test, preds),
        precision_score(y_test, preds),
        recall_score(y_test, preds),
        f1_score(y_test, preds),
        roc_auc_score(y_test, probs)
    ])
```

```python
# Save model evaluation result

results_df = pd.DataFrame(
    results,
    columns=["Segment", "Accuracy", "Precision", "Recall", "F1", "ROC_AUC"]
)

results_df.to_csv("model_evaluation_results.csv", index=False)
results_df
```

```python
# Hyperparameter Tuning - Gridsearch

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring="f1"
)

grid.fit(X_train, y_train)
grid.best_params_
```
